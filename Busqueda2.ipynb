```python
import pandas as pd
import re


def eliminar_espacios_adicionales(cadena):
    if(type(cadena) == float):
        return "NO"
    return re.sub(r'\s+', ' ', cadena).strip()

def transformar_string(texto):
    # Convertir a mayúsculas
    texto = texto.upper()
    # Reemplazar tildes por letras sin tilde
    texto = texto.replace('Á', 'A').replace('É', 'E').replace('Í', 'I').replace('Ó', 'O').replace('Ú', 'U')
    # Reemplazar la letra "ñ" por "n"
    texto = texto.replace('Ñ', 'N')
    return texto

def limpiar_texto(texto):
    # Eliminar caracteres no deseados y convertir a mayúsculas
    texto_limpio = re.sub(r'[^A-Z] ', '', texto.upper())
    return texto_limpio

def split_and_pad_names(name_string, target_length=5):
    parts = name_string.split()
    # Asegúrate de que la lista tenga la longitud deseada, completando con None si es necesario
    parts += [None] * (target_length - len(parts))
    return parts[:target_length] 
```

```python
datos = pd.read_csv(r'C:\Users\limc_\Downloads\padron_definitivo_nacional.txt', sep='|', encoding="latin")
datos
```

```python
datos2 = datos[datos.CALIDAD.isnull()]
datos2
```

```python
datos2["Nombre_merge"] = datos2["NOMBRE"].apply(eliminar_espacios_adicionales).apply(transformar_string).apply(limpiar_texto)
datos2
```

```python
nombre_counts = datos2["Nombre_merge"].value_counts()

# Función optimizada
def getNumeroCandidatos(nombre):
    return nombre_counts.get(nombre, 0)


#def getNumeroCandidatos(nombre):
#    return len(datos2[datos2["Nombre_merge"] == nombre])
```

```python
df = pd.read_excel(r"C:\Users\limc_\Downloads\Municipalidad de Angol.xlsx")
df
```

```python
df2 = df[["Nombres","Paterno","Materno"]].drop_duplicates()
df2["Nombres2"] = df2["Nombres"].apply(eliminar_espacios_adicionales).apply(transformar_string).apply(limpiar_texto)
df2["Paterno2"] = df2["Paterno"].apply(eliminar_espacios_adicionales).apply(transformar_string).apply(limpiar_texto)
df2["Materno2"] = df2["Materno"].apply(eliminar_espacios_adicionales).apply(transformar_string).apply(limpiar_texto)
df2[["n1","n2","n3","n4","n5"]] =df2["Nombres2"].apply(lambda x: split_and_pad_names(x)).apply(pd.Series)
df2 =df2.reset_index()
del df2["index"]
df3 = df2.copy()

#df4["cantidad"] = df4["NombreCompleto"].apply(lambda x: len(x.split(" ")))
df3["NombreCompleto"] = df3[['Paterno2', 'Materno2','n1', 'n2', 'n3', 'n4', 'n5',]].apply(
    lambda row: ' '.join(row.dropna().astype(str)).strip(), axis=1
)
df4 = df3[["NombreCompleto"]].drop_duplicates()
df4["Nombre_merge"] = df4["NombreCompleto"].apply(eliminar_espacios_adicionales).apply(transformar_string).apply(limpiar_texto).apply(lambda x: x.replace(" ","+"))
```

```python
df4["numero"] = df4["NombreCompleto"].apply(getNumeroCandidatos)
df4
```
