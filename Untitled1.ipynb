```python
import pandas as pd
import requests
import re

# Expresión regular para detectar cadenas que contienen al menos una letra
patron = re.compile(r'[a-zA-Z]')
```

```python
variable = pd.read_excel(url, sheet_name="Variables 2", skiprows=0)
```

```python
diccionarioMaster = {}
for i in variable.itertuples():
    diccionarioMaster[i[3]] =  i[2]
    #print(i[2],i[3])
diccionarioMaster["Codcom"] = "Codigo"
diccionarioMaster["Comuna"] = "Municipio"
```

```python
diccionarioAnual = {}
for i in range(2001,2024):
    diccionarioAnual[i] = []
```

```python
import os

# Especifica la ruta de la carpeta
carpeta = 'data2'

# Obtiene la lista de archivos y directorios en la carpeta
archivos_y_directorios = os.listdir(carpeta)

# Filtra solo los archivos (excluyendo los directorios)
archivos = [f for f in archivos_y_directorios if os.path.isfile(os.path.join(carpeta, f))]

# Imprime la lista de archivos
for archivo in archivos:
    anio = int(archivo.split("-")[2][:-5]) 
    diccionarioAnual[anio].append(f"data2/{archivo}")
```

```python
def getDFAnual(url,anio):
    df = pd.read_excel(url, header=None)
    # Asignar la segunda fila como encabezado
    df.columns = df.iloc[2]
    #del df["2"]
    #df.columns[1] = "Codcom"
    #df.columns[2] = "Comuna"
    # Eliminar la segunda fila (que ahora es el encabezado) y la tercera fila
    df = df.drop([0, 1, 2,3])
    # Reiniciar los índices del DataFrame
    df = df.reset_index(drop=True)
    df.columns = ["Codcom","Comuna"] + list(df.columns[2:])
    df["Año"] = anio
    return df
```
