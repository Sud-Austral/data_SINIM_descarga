```python
import pandas as pd
import xml.etree.ElementTree as ET

def clean_xml(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    
    # Eliminar líneas en blanco al principio del archivo
    while lines[0].strip() == '':
        lines.pop(0)
    
    # Unir las líneas en un solo string
    xml_string = ''.join(lines)
    return xml_string

def parse_spreadsheetml(xml_string):
    # Parsear el string XML
    root = ET.fromstring(xml_string)
    
    # Definir los namespaces usados en el archivo SpreadsheetML
    namespaces = {
        'ss': 'urn:schemas-microsoft-com:office:spreadsheet'
    }

    # Extraer datos de las filas y celdas
    data = []
    for row in root.findall('.//ss:Row', namespaces):
        row_data = []
        for cell in row.findall('ss:Cell', namespaces):
            data_element = cell.find('ss:Data', namespaces)
            if data_element is not None:
                row_data.append(data_element.text)
            else:
                row_data.append(None)
        data.append(row_data)

    # Crear un DataFrame con los datos extraídos
    df = pd.DataFrame(data)
    return df

# Ruta al archivo SpreadsheetML
#file_path = 'Descarga-1-2021.xls'  # Asegúrate de que es un archivo XML válido

# Limpiar y leer el archivo XML
#xml_string = clean_xml(file_path)

# Parsear el archivo y obtener el DataFrame
#df = parse_spreadsheetml(xml_string)
```

```python
def saveXLSX(fila):
    url = fila[5]
    id_area = fila[1]
    anio = fila[4]
    file_path = f"data/Descarga-{id_area}-{anio}"
    xml_string = clean_xml(file_path) 
    df = parse_spreadsheetml(xml_string)
    df.to_excel(f"data2/Descarga-{id_area}-{anio}.xlsx", index=False)
    
    
```
